```python
from nytimesarticle import articleAPI
api = articleAPI('3761b7d1f36742218377831a3e59edeb')
articles = api.search( q = "Tesla", 
                      fq = {'headline':'Tesla', 'source':['Reuters','AP', 'The New York Times']}, 
     begin_date = 20170303)
print(articles)
```

```python
import math
hits = articles['response']['meta']['hits']
print("hits :" + str(hits))
ls = []
pages = int(math.ceil(hits/10))
for i in range(pages):
    print("collecting page " + str(i))
    articles['page'] = i
    for article in articles['response']['docs']:
        ar = article['web_url'].encode('utf-8')
        if not ar.__contains__("video"):
            ls.append(article['web_url'].encode('utf-8'))
        else:
            continue
print(ls)
```

```python
from bs4 import BeautifulSoup
import requests
import unicodedata

output = ""

for x in ls[:]:
    page = requests.get(x)
    soup = BeautifulSoup(page.content, 'html.parser')
    soup = soup.find('article')
    for script in soup(["script", "style"]):
        script.extract() 
    text = soup.text.strip()
    lines = (line.strip() for line in text.splitlines())
    chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
    text = ' '.join(chunk for chunk in chunks if chunk)
    output1 = text.encode('utf-8').strip()
    output = output + " " + output1
```
